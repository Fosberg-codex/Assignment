# Computational Physics

Computational physics refers to the use of numerical analysis for solving physics problems. It is the first field to have utilized modern computers in scientific research, and is now considered a subset of computational science. While some view it as a subdiscipline of theoretical physics, others see it as an intermediary between theoretical and experimental physics, offering supplementary insights to both areas of study.

Computational physics pertains to numerical approximations that involve expressing the solution as a finite sequence of simple mathematical operations (an algorithm), which are then computed using a computer to obtain an approximate solution and the corresponding error.

It has applications in various fields of physics, such as astrophysics, condensed matter physics, particle physics, and plasma physics, among others. It has also played a crucial role in advancing scientific research and engineering design, especially in areas where theoretical calculations and experimental observations are difficult, expensive, or impossible to obtain.

# Link between the task given and computational physics

The solution of the time-independent Schrödinger equation for a particle in a one-dimensional potential well using numerical methods is a key application of computational physics.
In computational physics, the time-independent Schrödinger equation is a fundamental equation that describes the behavior of a quantum mechanical particle in a one-dimensional potential well. Solving this equation involves finding the energy eigenvalues and eigenfunctions of the system, which are important quantities that determine the particle's behavior in the well.
However, in most cases, the time-independent Schrödinger equation cannot be solved analytically, especially for complex potential wells. Therefore, numerical methods are used to obtain approximate solutions. These numerical methods involve expressing the differential equation as a discrete set of algebraic equations that can be solved using computer algorithms.

#### importing all python libraries relevant for the task



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig
```

#### Building fundamental models in classes

```python

class ParticleInABox:
    def __init__(self, L, m, V):
        self.m = m  # mass of the particle
        self.V = V  # potential energy function
        self.L = L  # length of the box
        
        
class ParticleInABoxSolver:
    def __init__(self, particle, N):
        self.particle = particle
        self.N = N
        self.hbar = 1.055e-34  # Planck's constant divided by 2π
        self.dx = particle.L / N  # step size
        self.x = np.linspace(0, particle.L, N)  # position arra
        self.E = None  # energy levels
        self.psi = None  # wave functions
        self.H = None  # Hamiltonian matrix
        
        
    def set_hamiltonian_matrix(self):
        self.H = np.zeros((self.N, self.N))
        for i in range(1, self.N-1):
            self.H[i, i] = -2 / (self.dx ** 2) + self.particle.V(self.x[i])
            self.H[i, i-1] = 1 / (self.dx ** 2)
            self.H[i, i+1] = 1 / (self.dx ** 2)
        self.H[0, 0] = -2 / (self.dx ** 2) + self.particle.V(self.x[0])
        self.H[0, 1] = 1 / (self.dx ** 2)
        self.H[-1, -1] = -2 / (self.dx ** 2) + self.particle.V(self.x[-1])
        self.H[-1, -2] = 1 / (self.dx ** 2)
        
    def calc(self):
        self.set_hamiltonian_matrix()
        self.E, self.psi = eig(self.H)
        idx = self.E.argsort()
        self.E = self.E[idx]
        self.psi = self.psi[:, idx]
        
    def plot_wave_functions(self, num_functions=5):
        fig, ax = plt.subplots()
        for i in range(num_functions):
            ax.plot(self.x, self.psi[:, i], label=f"n={i+1}")
        ax.set_xlabel("x")
        ax.set_ylabel("ψ(x)")
        ax.legend()
        plt.show()
        
    def plot_energy_levels(self):
        fig, ax = plt.subplots()
        ax.plot(range(1, len(self.E)+1), self.E, 'bo')
        ax.set_xlabel("n")
        ax.set_ylabel("Energy (J)")
        ax.set_title("Particle in a one-dimensional box")
        plt.show()

```

#### Visualising the relationships

```python
V = lambda x: np.inf if x < 0 or x > 1 else 0

particle = ParticleInABox(1, 1, V)
solver = ParticleInABoxSolver(particle, N=100)
solver.calc()
solver.plot_energy_levels()
solver.plot_wave_functions(num_functions=5) 
```
